```python
from collections import defaultdict
class Graph:
    def __init__(self):
        self.graph = defaultdict(list)
    def addEdge(self,u,v):
        self.graph[u].append([v,0])
        self.graph[v].append([u,0])
    def printgraph(self):
        print(self.graph)
    def dfs(self, start, goal, i):
        opened,closed = [],[]
        opened.append([start, 0])
        while opened:
            u,v = opened.pop(0)
            closed.append(u)
            if u == goal: print(u,' ',opened,' ', closed); return 1
            t_li =[]
            if v<i :
                for j in self.graph[u]:
                    j[1]=v+1
                    if j[0] not in closed: t_li.append(j)
                opened = t_li + opened
            print(u,' ',opened,' ', closed)
    def dfid(self, start,goal):
        i = x = 0
        if goal not in self.graph:
            print("Goal can't be found")
            return
        while x!=1:
            print("Depth = ", i)
            x = self.dfs(start, goal, i)
            i+=1
        print('Goal found')
g = Graph()
n = int(input())
for i in range(n):
    u,v = input().split()
    g.addEdge(u,v)
start,goal = input().split()
g.dfid(start,goal)
#g.printgraph()
```

```python
class WaterJug:
    def __init__(self,bj_max,sj_max,bj,sj,goal):
        self.bj_max = bj_max
        self.sj_max = sj_max
        self.bj = bj
        self.sj = sj
        self.goal = goal
    
    def fillbj(self):
        self.bj = self.bj_max
        print("(",self.bj,",",self.sj,")")
    
    def fillsj(self):
        self.sj = self.sj_max
        print("(",self.bj,",",self.sj,")")
        
    def emptybj(self):
        self.bj = 0
        print("(",self.bj,",",self.sj,")")
    
    def emptysj(self):
        self.sj = 0
        print("(",self.bj,",",self.sj,")")
    
    def transferbjtosj(self):
        while True:
            self.bj = self.bj-1
            self.sj = self.sj+1
            
            if self.bj == 0 or self.sj == self.sj_max:
                break
        print("(",self.bj,",",self.sj,")")
        
    def transfersjtobj(self):
        while True:
            self.sj = self.sj-1
            self.bj = self.bj+1
            
            if self.sj == 0 or self.bj == self.bj_max:
                break
        print("(",self.bj,",",self.sj,")")
        
    def measuregoal_bs(self):
        print("(",self.bj,",",self.sj,")")
        
        while(True):
            
            if self.bj == self.goal or self.sj == self.goal:
                print("Success")
                break
            
            if self.bj==0:
                self.fillbj()
            elif self.bj > 0 and self.sj < self.sj_max:
                self.transferbjtosj()
            elif self.bj > 0 and self.sj == self.sj_max:
                self.emptysj()

    def measuregoal_sb(self):
        print("(",self.bj,",",self.sj,")")
        
        while(True):
            if self.bj == self.goal or self.sj == self.goal:
                print("success")
                break
        
            if self.sj==0:
                self.fillsj()
            elif self.sj>0 and self.bj<self.bj_max:
                self.transfersjtobj()
            elif self.sj>0 and self.bj==self.bj_max:
                self.emptybj()
        
if __name__ == '__main__':
    wj = WaterJug(5,3,0,0,4)
    wj.measuregoal_bs()
    wj1 = WaterJug(5,3,0,0,4)
    wj1.measuregoal_sb()
```

```python
class BFS_algo:
    def __init__(self, graph):
        self.graph = graph
    
    def bfs(self,s):
        op = [s]
        cl = []
        while len(op)>0:
            t = op.pop(0)
            for i in self.graph[t]:
                if i not in cl and i not in op and i != t:
                    op.append(i)
            cl.append(t)
        print(cl)
            
    def bfs_goal(self,s,g):
        op = [s]
        cl = []
        while len(op)>0:
            t = op.pop(0)
            if t==g:
                print("goal")
                break
            for i in self.graph[t]:
                if i not in cl and i not in op and i != t:
                    op.append(i)
            cl.append(t)
        #print(op)
        print(cl)
        print(g)


if __name__ == '__main__':
    #graph = {'A': ['B','C'],'B':['D','E'],'C':['F','G'],'D':[],'E':[],'F':[],'G':[]}
    graph = {0:[1,2],1:[2],2:[0,3],3:[3]}
    t1 = BFS_algo(graph)
    t1.bfs(0)
    t1.bfs_goal(0,2)
    
```

```python
class DFS_algo:
    def __init__(self,graph):
        self.graph = graph
    
    def dfs(self,s):
        op = [s]
        cl = []
        while len(op)>0:
            t = op.pop(0)
            t_li = []
            for i in self.graph[t]:
                if i not in op:
                    t_li.append(i)
            #t_li.reverse()
            op = t_li + op
            cl.append(t)
        print(cl)
        
if __name__=='__main__':
    graph = {'A': ['B','C'],'B':['D','E'],'C':['F','G'],'D':[],'E':[],'F':[],'G':[]}
    d = DFS_algo(graph)
    d.dfs('A')
```
